```python
input_1 = pd.read_excel('/kaggle/input/future-skills/Data_1.xlsx')
input_2 = pd.read_excel('/kaggle/input/future-skills/Data_2.xlsx')
```

```python
input_1
```

```python
merged_data = pd.merge(input_1, input_2, left_on="User ID", right_on="uid")
merged_data['Team Name'] = merged_data['Team Name'].str.lower() 
```

```python
team_avg = merged_data.groupby('Team Name').agg({'total_statements': 'mean', 'total_reasons': 'mean'}).reset_index()
```

```python
team_avg.round(2)
```

```python
team_avg['Rank'] = team_avg['total_statements'] + team_avg['total_reasons']
team_avg = team_avg.sort_values(by='Rank', ascending=False).reset_index(drop=True)
team_avg['Rank'] = team_avg.index + 1
```

```python
team_avg
```

```python
team_avg_df = team_avg.rename(columns={'total_statements': 'Average Statements', 'total_reasons': 'Average Reasons', 'Team Name': 'Thinking Teams Leaderboard' , 'Rank' : 'Team_Rank'})
```

```python
team_avg_df
```

```python
team_avg_df["Thinking Teams Leaderboard"] = team_avg_df["Thinking Teams Leaderboard"].apply(lambda x: "".join(word.title() for word in x.split()))
```

```python
team_avg_df.set_index("Team_Rank", inplace=True)
```

```python
team_avg_df
```

```python
team_avg_df.to_excel('output1.xlsx' , index = False)
```

```python
merged_data = merged_data.sort_values(by=["total_statements", "total_reasons"], ascending=False)
```

```python
rank_df = pd.DataFrame({
    'Rank': (merged_data["total_statements"] + merged_data["total_reasons"]).rank(method='min', ascending=False).astype(int)
})
```

```python
merged_df = pd.concat([merged_data, rank_df], axis=1)
```

```python
output_df = merged_df.loc[:, ["Rank", "name", "uid", "total_statements", "total_reasons"]]
```

```python
output_df.set_index("Rank", inplace=True)
```

```python
output_df["name_lower"] = output_df["name"].str.lower()
```

```python
output_df["name_lower"] = output_df["name_lower"].str.replace('[^A-Za-z\s]+', '')
```

```python
output_df = output_df.sort_values(by=["Rank", "name_lower"], ascending=[True, True])
```

```python
output_df['name_lower']
```

```python
output_df = output_df.drop('name_lower', axis=1)
```

```python
output_df = output_df.rename(columns={'Rank' : 'Actual_Rank','name': 'Name', 'uid': 'UID', 'total_statements': 'No. of Statements', 'total_reasons' : 'No. of Reasons'})
```

```python
output_df['Increasing_order_Rank'] = output_df.reset_index().index + 1
```

```python
output_df
```

```python
output_df.to_excel('output2.xlsx' , index = False)
```
